```python
import pandas as pd
import numpy as np
pd.set_option('display.max_colwidth', -1)
```

```python
data = pd.read_csv('../../dataset/merged_465_entity_uri_google_pre_seq_0_50.csv', encoding='utf-8', usecols=['sequence1', 'sequence2'])
data.info()
```

```python
display(data.head(50))
len(data['sequence1'].iloc[0:10].values)
```

```python
print (sum(data.duplicated(subset=['sequence1'])))
# train, dev, test ratios
total = len(data.index)
print (type(total))
train_size = int(70*total/100)
dev_size= int(10*total/100)
test_size = int(20*total/100)
print (total, train_size, dev_size, test_size)

train_index = [0, train_size]
dev_index = [train_size, train_size+dev_size]
test_index = [train_size+dev_size, train_size+dev_size+test_size]
print (train_index, dev_index, test_index)
```

```python
indexs = []
indexs.append(train_index)
indexs.append(dev_index)
indexs.append(test_index)

print (indexs)
dataFilesuffix =['train_glove_0_50', 'dev_glove_0_50', 'test_glove_0_50'] 
for suf, index in zip(dataFilesuffix, indexs):
    X1 = data['sequence1'].iloc[index[0]:index[1]].values
    X2 = data['sequence2'].iloc[index[0]:index[1]].values
    print (len(X1), len(X2))
    np.savetxt('../../dataset/trainingData/'+suf+'.seq1', encoding='utf-8', X=X1, fmt='%s')
    np.savetxt('../../dataset/trainingData/'+suf+'.seq2', encoding='utf-8', X=X2, fmt='%s')
```
