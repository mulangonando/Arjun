```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

```python
def extractLoss(line):
    spt = line.strip().split(':')
    return spt[-1]
    

with open('../Seq2SeqPyTorch/En_DeAttention_10Nov.log') as f:
    lines = f.readlines()
    loss =  [float(extractLoss(line)) for line in lines if 'Average loss' in line ]
    #print (loss)

#plt.plot(loss)
def showPlot(points):
    plt.figure()
    fig, ax = plt.subplots()
    # this locator puts ticks at regular intervals
    loc = ticker.MultipleLocator(base=0.3)
    ax.yaxis.set_major_locator(loc)
    plt.plot(points)

showPlot(loss)
```
